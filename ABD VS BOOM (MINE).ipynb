```python
jassi_df.groupby(["striker"])["runs_off_bat"].sum().sort_values(ascending=False).head(10)
```

```python
jassi_df1=pd.DataFrame(jassi_df.groupby(["striker"])["runs_off_bat"].sum()).reset_index()
```

```python
jassi_df2=pd.DataFrame(jassi_df.groupby(["striker"])["ball"].count()).reset_index()
```

```python
jassi_df1
```

```python
jassi_df2
```

```python
jassi_df3=jassi_df1.merge(jassi_df2,on="striker",how="left")
```

```python
jassi_df3
```

```python
jassi_df3["strike_rate"]=100*jassi_df3["runs_off_bat"]/jassi_df3["ball"]
```

```python
jassi_df3.head(5)
```

```python
# lets put a minimum criateria of 30 balls 
```

```python
jassi_df3=jassi_df3[jassi_df3["ball"]>=30]
```

```python
abd=df[df["striker"]=="AB de Villiers"]
```

```python
abd1=pd.DataFrame((abd.groupby("bowler")["runs_off_bat"]).sum()).reset_index()
```

```python
abd1
```

```python
abd1.merge(abd2,on="bowler",how="left")
```

```python
abd3=abd1.merge(abd2,on="bowler",how="left")
```

```python
plt.figure(figsize=(16,8))
plt.scatter(jassi_df3.strike_rate,jassi_df3.runs_off_bat)
# plt.text(x,y,text)
for i in range(len(jassi_df3)):
    plt.text(jassi_df3["strike_rate"][i]-5,jassi_df3["runs_off_bat"][i]+1,jassi_df3["striker"][i])
plt.axvline(130,ls="--",color="red")
plt.axhline(70,ls="--",color="grey")
plt.xlabel("strike_rate",size=15)
plt.ylabel("Runs",size=15)
plt.title("Batsmen vs Bumrah",size=20)
plt.show()
```
